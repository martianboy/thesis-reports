#توزیع تصادفی دیریکله
##برای مدل‌سازی توابع جرم احتمال

یک تاس شش‌سویه را می‌توان نمونه‌ای از یک تابع جرم احتمال (که از این پس به اختصار pmf خواهیم نامید) دانست. برای نمونه‌گیری از این pmf، تاس می‌اندازیم و عددی بین یک تا شش می‌بینیم. اما قوانین فیزیک و شرایط تولید در کارخانه‌ها موجب می‌شود تا تاس‌های واقعی کاملا یک‌نواخت نباشند. یک کیسه‌ی صدتایی از تاس‌ها، نمونه‌ای از یک pmf تصادفی است. برای نمونه‌گیری از این متغیر تصادفی، دست در کیسه برده و یک تاس را بیرون می‌آوریم؛ یعنی، یک pmf نمونه انتخاب می‌کنیم. می‌توان انتظار داشت یک کیسه از تاس‌ها که با روش‌های قدیمی یک قرن پیش ساخته شده، به شدت نسبت به pmf یک‌نواخت انحراف داشته باشند؛ اما در یک کیسه از تاس‌های آخرین مدلی که در قمارخانه‌های لاس وگاس مورد استفاده قرار می‌گیرد، به سختی می‌توان انحراف قابل لمس پیدا کرد. تصادفی‌بودن توابع جرم احتمال را می‌توان با «توزیع دیریکله» مدل کرد.

کاربرد توزیع دیریکله به خصوص در زمینهٔ مدل‌سازی توزیع واژگان در اسناد متنی مفید واقع شده است. اگر یک واژه‌نامه شامل k واژه داشته باشیم، در این صورت یک سند متنی را می‌توان به صورت یک pmf به طول k نمایش داد که احتمال هر جزء یا واژه، نسبت تعداد تکرار آن واژه به تعداد کل واژگان موجود در آن سند خواهد بود. یک مجموعه از اسناد متنی، دسته‌ای از pmfها را تشکیل می‌دهند و ما با استفاده از توزیع دیریکله، می‌توانیم پراکندگی آن‌ها را مدل‌سازی کنیم. توزیع‌های دیریکله را می‌توان برای مدل‌سازی اسناد بر اساس نویسندگان یا موضوعات مختلف به کاربرد.

در ابتدا به توزیع دیریکله و برخی از خصوصیات جالب توجه آن می‌پردازیم.

## تعریف توزیع دیریکله

یک pmf با $k$ جزء را می‌توان به صورت نقطه‌ای روی سیمپلکس $k-1$ بعدی احتمال در نظر گرفت. این سیمپلکس که با $\Delta_k$ نمایش داده می‌شود، مجموعه‌ی بردارهایی با عناصر نامنفی است که جمع عناصرشان برابر یک باشد؛ یعنی: $\Delta_k=\left\{q\in\mathbb{R}:\sum^k_{i=1}{q_i}=1,\,q_i\geq 0 \;\text{for}\;i\in\{1,\dots,k\}\right\}$. با این که $\Delta_k$ در فضای $k$ بعدی قرار می‌گیرد، خود یک شیء $k-1$ بعدی است. هر نقطه $q$ روی این سیمپلکس به خودی خود می‌تواند یک تابع جرم احتمال در نظر گرفته شود؛ چرا که هر جزء $q$ نامنفی است و جمع تمام اجزاء برابر یک خواهد بود. توزیع دیریکله را می‌توان یک توزیع احتمالی روی سیمپلکس احتمالی $k-1$بعدی $\Delta_k$ در نظر گرفت؛ یعنی توزیعی روی همه‌ی pmfها از طول k.

**توزیع دیریکله:** فرض کنیم $Q=\left[Q_1,\dots,Q_k\right]$ یک pmf تصادفی باشد؛ یعنی یک متغیر تصادفی که هر نمونه از آن یک pmf است. همچنین فرض کنیم $\alpha=\left[\alpha_1,\dots,\alpha_k\right]$ با شرط $\alpha_i>0$ به ازای هر $i$ باشد و گیریم $\alpha_0=\sum^k_{i=1}{\alpha_i}$. در این صورت گوییم $Q$ دارای توزیع دیریکله با پارامتر $\alpha$ است و آن را با $Q\sim\text{Dir}\left(\alpha\right)$ نشان می‌دهیم، هرگاه اگر $q$ یک pmf نباشد، داشته باشیم $f(q;\alpha)=0$ و اگر $q$ یک pmf باشد، آنگاه:

$$
\newcommand\given[1][]{\:#1\vert\:}
f(q;\alpha)=\frac{\Gamma\left(\alpha_0\right)}{\prod^k_{i=1}{\Gamma\left(\alpha_i\right)}}\prod^k_{i=1}{q_i^{\alpha_i-1}}
$$

در شکل‌های زیر، از توزیع دیریکله با مقادیر مختلف بردار $\alpha$ تعداد ۴۰۰ نمونه گرفته شده است. دیده می‌شود که اگر همه ی عناصر $\alpha$ برابر باشند، تمرکز نمونه‌ها حول مرکز خواهد بود. اگر عناصر کمتر از یک باشند، نمونه‌ها از مرکز دور و به سمت حاشیه سیمپلکس می‌روند؛ برعکس هر چه عناصر $\alpha$ از یک بزرگ‌تر باشند، تمرکز دور نقطه مرکزی چگال‌تر است. $\alpha$یی که تمام عناصرش برابر یک باشد، نمونه‌ها را به طور یک‌نواخت در سطح سیمپلکس پراکنده می‌کند.

```python
from mpl_toolkits.mplot3d import Axes3D

def set_lim(ax, *lim):
    ax.set_xlim(*lim)
    ax.set_ylim(*lim)
    ax.set_zlim(*lim)
```

```python
fig = plt.figure(figsize=(13,11))

n = 400
axes = [fig.add_subplot(2,2,i+1, projection='3d') for i in range(4)]
alpha_vectors = [(1,1,1), (.1,.1,.1),(10,10,10),(2,5,15)]

for i in range(4):
    s = np.random.dirichlet(alpha_vectors[i], n).transpose()
    axes[i].scatter(*s, zdir='z')
    axes[i].set_title('alpha = {0}'.format(alpha_vectors[i]))
    set_lim(axes[i], 0, 1)
    axes[i].view_init(30, 45)

plt.show()
```

به ازای $k=2$ به آسانی دیده می‌شود که توزیع دیریکله به توزیع بتا تبدیل می‌شود. توزیع بتا $\text{Beta}\left(\alpha,\beta\right)$ روی بازه‌ی $(0,1)$ تعریف می‌شود و دارای تابع چگالی به صورت زیر است:

$$
f(x;\alpha,\beta)=\frac{\Gamma(\alpha+\beta)}{\Gamma(\alpha)\Gamma(\beta)}x^{\alpha-1}{\left(1-x\right)}^{\beta-1}
$$

یعنی اگر $X \sim \text{Beta}(a,b)$ باشد، در این صورت $Q = (X,1-X) \sim \text{Dir}(\alpha)$ خواهد بود که در آن $\alpha=[a,b]$ و برعکس.

در جدول زیر خصوصیات توزیع تصادفی دیریکله را به طور خلاصه مشاهده می‌کنید:

|                    |                                                                                          |
|--------------------|-----------------------------------------------------------------------------------------:|
| چگالی              |                                   $\frac{1}{B(\alpha)}\prod^{d}_{j=1}{q_j^{\alpha_j-1}}$ |
| امید ریاضی         |                                                              $\frac{\alpha}{\alpha_0}$ |
| کوواریانس          |           For $i\ne j$, $Cov(Q_i,Q_j)=\frac{-\alpha_i\alpha_j}{\alpha_0^2(\alpha_0+1)}$ |
| واریانس              | for all i, $Var(Q_i)=\frac{\alpha_i(\alpha_0-\alpha_i)}{\alpha_0^2(\alpha_0+1)}$ |
| مد                 |                                                             $\frac{\alpha-1}{\alpha0-k}$ |
| توزیع‌های حاشیه‌ای |                                         $Q_i\sim\text{Beta}(\alpha_i,\alpha_0-\alpha_i)$ |
| توزیع شرطی         |   $\left(Q_{-i}\given[\big] Q_i\right)\sim\left(1-Q_i\right)\text{Dir}\left(\alpha_{-i}\right)$ |

## مزدوج پیشین توزیع چندجمله‌ای

یک توزیع چندجمله‌ای با پارمتر طبیعی $n$ و یک pmf مانند $q = \left[q_1, q_2, \dots, q_k\right]$ به صورت مشخص می‌شود: اگر $n$ پیشامد مستقل داشته باشیم و در هر پیشامد، احتمال دیدن خروجی $i$ام $q_i$ باشد، در این صورت توزیع چندجمله‌ای احتمال دیدن خروجی $i$ام به تعداد $x_i$بار را تعیین می‌کند. مثلا در یک تاس $k$ وجهی، $x_i$ تعداد دفعاتی که وجه $i$ام دیده می‌شود را نشان می‌دهد. اگر داشته باشیم $X\sim \text{Multinomial}_k\left(n,q\right)$، در این صورت تابع جرم احتمال آن به صورت زیر خواهد بود:

$$
f\left(x_1,x_2,\dots,x_k\given[\Big] n,q=\left(q_1,q_2,\dots,q_k\right)\right) = \frac{n!}{x_1!x_2!\dots x_k!}\prod_{i=1}^{k}{q_i^{x_i}}
$$

توزیع دیریکله را می‌توان به عنوان یک مزدوج پیشین برای پارامتر $q$ در توزیع چندجمله‌ای به کار برد؛ یعنی، اگر $\left(X\given[\big]q\right)\sim\text{Multinomial}_k\left(n,q\right)$ باشد و $Q\sim\text{Dir}\left(\alpha\right)$، در این صورت داریم: $\left(Q\given[\big]X=x\right)\sim\text{Dir}\left(\alpha+x\right)$. از جمله مزایای استفاده از توزیع دیریکله بدین منظور، آن است که می‌توان توزیع پسین $Q$ را به سادگی محاسبه کرد و همچنین به طریقی، فهمید پس از جمع آوری داده‌ها، چقدر باورهای ما تغییر یافته است.
